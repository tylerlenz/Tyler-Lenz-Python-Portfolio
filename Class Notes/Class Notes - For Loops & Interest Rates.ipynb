# For Loops

The structure of for loops:

* for i in range(n)
* for i in range(start, end + 1, interval)

See examples:

```python
# example of a basic for loop

for i in range(1, 25):
    print(i, end = ", ")
```

As you can see, this for loops just counts and prints our entire range one-by-one. Now, lets try using a for loop with a list.

```python
range_list = list(range(25,201))
for i in range_list:
    print(i, end = ", ")
```

Now, we can make a for loop that counts backwards using lists.

```python
lst = []
# create list of numbers from 0 to 100
for i in range(25):
    lst.insert(i, i)
# You can call the value at some index (location) in the list by calling 
# lst[#]
print(lst[1])
print(lst[10])
# identify very last index with -1
print(lst[-1], lst[24])
print(lst[-2], lst[23])
print(lst[-3], lst[22])
```

```python
# Create a list of numbers that counts backward from 100 to 0
lst2 = []
for i in range(76):
    lst2.insert(0, i)
print(lst2)
```

We can also transform dictionaries to pandas dataframes.

Dictionaries of two forms can be transformed into pandas dataframes. A dictionary can have 1 layer of keys, each of which point to a list:

## Calculating Interest Rates

We can calculate interest rate values using python.

$FV = PV(1+r)^t$

$r = (\frac{FV}{PV})^{1/t}-1$

Suppose that three years after you place 1500 dollars in a deposit account, the value of the deposit grows to 2000 dollars. Build python script that calculates the annual interest rate.

```python
FV = 2000
PV = 1500
t = 3
r = (FV / PV) ** (1 / t) - 1
r
```

Which, is the interest rate. To check that our calculations are correct, we can put our interest rate back into the future value function to make sure we got a correct answer.

```python
FV_test = PV * (1 + r) ** t
FV_test
```

Interest can be compounded annually (once per year) or it can be compounded more than once. If the interest is compounded more than once annually, in the present equation we divide the interest by the number times compounded, $n$, and we multiply the number of years, $t$, by $n$.

$FV = PV(1+\frac{r}{n})^{tn}$

Next, we will create a dictionary that records the value of $2000 after a year of collecting interest over different rates of compounding.

```python
# use dictionary to define how many times interest is compounded
compounding_def = {"annual":1,
                  "semi-annual":2,
                  "quarterly":4,
                  "monthly":12,
                  "weekly":52,
                  "daily":365}

PV = 2000
r = 0.1
t = 1
#compound interest at rate of 10%
# 10% == 0.1
# create dictionary to hold the results
compounding_interest = {}

# for var1, var2 in dict.items()
# for key1, obj1 in dict.items()
for period_name, n in compounding_def.items():
    # use round(val, #decimals) to round value to a specified # of decimals
    compounding_interest[period_name] = round(PV * (1 + r / n) ** (t * n), 2)
compounding_interest
```

```python
compounding_def = {"annual":1,
                   "semi-annual":2,
                   "quarterly":4,
                   "monthly":12,
                   "weekly":52,
                   "daily":365
                  }
# create variables
PV = 2000
t = 2
# use generator function to create list of interest rates
# from 1 to 20
rates = [i for i in range(1,21)]
compounding_interest = {}
for rate in rates:
    r = rate / 100
    rate = str(rate)+"%"
    compounding_interest[rate] = {}
    for period_name, n in compounding_def.items():
        compounding_interest[rate][period_name] = round(PV * (1 + r / n) ** (t * n), 2)
compounding_interest

#translating into a dictionary

compounding_interest_df = pd.DataFrame(compounding_interest)
compounding_interest_df
```

## Using Computational Methods to Brute Force Solve for a Value

We will be using conditional statements to solve for values.

```python
x = True
if x:
    print(x)
else:
    print("x is not True, x is equal to",x)
```

```python
for x in range(5):
    print("x ==", x)
    print("x % 2 ==", x % 2)
    if x % 2 == 0:
        print("x is an even number")
    else:
        print("x is not an even number")
```

Lastly, we will use the in class example, where we are using a margin value in order to find the solution to an equation.

```python
a = 4
x = 1
ax = a * x
# solve for ax = 1000
margin = 1.0


# create variable to track if ax is ablove or below 1000
below_1000 = True
decrease_margin = False

while round(ax,10) != 1000:
    if decrease_margin:
        margin = margin * .1
        decrease_margin = False
    # increase x or decrease x depending on whether
    # x is below or above 1000
    x = x + margin if ax < 1000 else x - margin
    ax = a * x
    print("x =", x, "\tax", ax)
    # if ax crosses over 1000, identify
    # 1) that ax is now greater than 1000
    # 2) x should change by a smaller margin and in a diff
    # direction
    if below_1000 == True and ax > 1000:
        #change by smaller margin
        decrease_margin = True
        # move in opposite direction
        below_1000 = False
    elif below_1000 == False and ax < 1000:
        # change by smaller margin
        decrease_margin == True
        # move in opposite direction
        below_1000 = True
    
print("Solution: (" + str(a)+")" + "("+str(x)+") =", ax)
```
