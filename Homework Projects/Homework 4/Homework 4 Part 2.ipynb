For this assignment, I will be utilizing the nominal gdp, real gdp, CPI and analyzing correlation. Then, I will calculate the year-over-year percent change and once again check the correlation.

```python
from datlib.plots import * 
from datlib.FRED import *

data_codes = {"GDP: Real Gross Domestic Product" : "GDPC1",
             "GDP: Gross Domestic Product" : "GDP",
             "Consumer Price Index: Total All Items for the United States" : "CPALTT01USQ657N",
             "Currency: Currency in Circulation" : "CURRCIR"}

start = datetime.datetime(1960, 1, 1)
end = datetime.datetime.today()

data = gather_data(data_codes = data_codes, start = start, end = end, freq = "Q")

data
```

```python
import seaborn as sn

corrMatrix = data.corr()
print (corrMatrix)
sn.heatmap(corrMatrix, cmap = "Spectral", annot=True)
plt.show()
```

Here is our correlation matrix. From it, we can see that Gross & Real GDP are positively correlated, whereas CPI and Real GDP are negatively correlated. The heatmap has a key so the more towards purple equals more positively correlated, and the dark red/maroon is more negatively correlated. Now, I will make a table of yearly percent changes for each variable. Then, analyze their correlation in the same way, with the same legend as above.

```python
change = data.pct_change()
print(change)
change.corr()
corrMatrix_2 = change.corr()
print(corrMatrix_2)
sn.heatmap(corrMatrix_2, cmap = "seismic", annot=True)
plt.show()
```

Once again, we can use our legend to view the correlations. The only yearly changes that are positively correlated (strongly, at least) are GDP and Real GDP. No variables have strong negative correlation, but a couple have a weak negative correlation. These will be the darker blue boxes. As we can see, there was more positive correlation between the variables themselves, not the year-by-year percent change.
